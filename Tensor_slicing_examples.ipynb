# Indexing 

- starts at zero 
- (-)ve indices count backwards from the end
- colons ":" are used for slice:start:stop:step


```python
rank1_tensor = tf.constant([11, 22, 33, 44, 55, 66, 77, 88, 99])
rank1_tensor
```

```python
#get just number by using .numpy()
rank1_tensor[0].numpy()
```

```python
# -1 is last entry, counting backwards
rank1_tensor[-8]
```

```python
#no need to type 0 if you start at index[0]
rank1_tensor[:5]
```

```python
#if you start in th middle type start and stop index
rank1_tensor[3:5]
```

```python
#get reverse tensor
rank1_tensor[::-1].numpy()
```

Multi-axes indexing

```python
rank2_tensor = tf.constant([[1, 2, 3],
                            [4, 5, 6],
                            [7, 8, 9]])
rank2_tensor
```

```python
# x[row_start:row_end, col_start:col_end]
rank2_tensor[1, 1]
```

```python
#x[row_start:row_end+1, col_start:col_end+1]
rank2_tensor[0:2, 0:2]

#same as

rank2_tensor[:2, :2]
```

```python
# just ":" means all entries, -1 is last row
rank2_tensor[-1, :].numpy()
```

```python
rank2_tensor[:, -1].numpy()
```

```python
rank2_tensor[-2, -1]
```

```python
  # ends at the end
  rank2_tensor[1:, 1:]
```

Rank3 Tensor

```python
rank3_tensor = tf.constant([
                            [[1, 2, 3, 4],
                             [4, 5, 6, 4]],
                            
                            [[10, 20, 30, 5],
                             [40, 50, 60, 5]],
                            
                            [[11, 22, 33, 6],
                             [44, 55, 66, 6]]
                            ])
rank3_tensor
#shape = stacks, row/stack, column/stack
```

```python
rank3_tensor[:, :, 0]
```

```python
rank3_tensor[:2, :, 1:3]
```
